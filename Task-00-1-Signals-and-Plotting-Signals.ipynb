```python
import numpy as np
import matplotlib.pyplot as plt
```

A sinusoidal can be written as:
$$x(t) = Asin(2 \pi ft + phase)$$

```python
t = np.linspace(0, 2.0, 80)
```

**Create a Sinusoidal Signal $x(t)$ and 3 different plotting in subplots**

```python
def make_sinusoidal(amp, freq, phase = 0):
    return amp * np.sin(2.0 * np.pi * freq * t + phase)
```

```python
x_t = make_sinusoidal(1,1)

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(11, 11))
fig.suptitle('Different plottings of a sinusoidal signal', fontsize=15)
# fig.

ax[0, 0].plot(t, x_t)
ax[0, 0].set_title('continuous plot', fontsize = 15)

ax[0, 1].stem(t, x_t)
ax[0, 1].set_title('stem plot', fontsize = 15)

ax[1,0].scatter(t, x_t)
ax[1,0].set_title('scatter plot', fontsize = 15)

ax[1,1].axis('off')

for a in ax.flat:
    a.set(xlabel='t', ylabel='x(t)')

plt.show()
```

**Create 3 Sinusoidal Signal $x1(t), x2(t) and x3(t)$ then combine and plot the signal**

```python
x1 = make_sinusoidal(1, 2)
x2 = make_sinusoidal(2, 3)
x3 = make_sinusoidal(3, 4)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15, 14))
fig.suptitle('3 Sinusoidal signal and a combined signal', fontsize=15)

ax[0, 0].plot(t, x1, 'r')
ax[0, 0].set(xlabel = 't', ylabel = 'x1(t)')

ax[0, 1].plot(t, x2, 'g')
ax[0, 1].set(xlabel = 't', ylabel = 'x2(t)')

ax[1, 0].plot(t, x3, 'b')
ax[1, 0].set(xlabel = 't', ylabel = 'x3(t)')

ax[1 ,1].plot(t, x1 + x2 + x3)
ax[1, 1].set(xlabel = 't', ylabel = 'x1(t) + x2(t) + x3(t)')

plt.show()
```
