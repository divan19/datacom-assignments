```python
import numpy as np
import matplotlib.pyplot as plt
```

$$x(t) = A\sin(2\pi f t + phase)$$

```python
def make_sinusoidal(time_points, amp, freq, phase = 0):
    return amp * np.sin(2.0 * np.pi * freq * time_points + phase)
```

```python
# time points
tpoints = np.linspace(0.0, 2.0, 20000)
```

```python
x1 = make_sinusoidal(tpoints, 1, 5)
x2 = make_sinusoidal(tpoints, 1, 10)
x3 = make_sinusoidal(tpoints, 1, 15)

# y is the composite signal achieved by combining above three signals
y = x1 + x2 + x3
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(11, 11))
fig.suptitle('3 Sinusoidal Signals and 1 Composite Signal', fontsize=15)

ax[0,0].plot(tpoints, x1)
ax[0,0].set_title('Signal-1', fontsize = 15)

ax[0,1].plot(tpoints, x2)
ax[0,1].set_title('Signal-2', fontsize = 15)

ax[1,0].plot(tpoints, x3)
ax[1,0].set_title('Signal-3', fontsize = 15)

ax[1,1].plot(tpoints, y)
ax[1,1].set_title('Signal-1 + Signal-2 + Signal-3', fontsize = 15)

plt.show()
```

```python
# fs represents sampling frequency
fs = 10            # 10 samples/second
step = 1.0/fs      # sampling interval in time domain to obtain fs
sample_points = np.arange(0, 2.0, step)
y = make_sinusoidal(sample_points, 1, 5) + make_sinusoidal(sample_points, 1, 10) + make_sinusoidal(sample_points, 1, 15)
plt.title('fs = 10', fontsize = 15)
plt.stem(sample_points, y)
plt.show()

fs = 20
step = 1.0/fs
sample_points = np.arange(0, 2.0, step)
y = make_sinusoidal(sample_points, 1, 5) + make_sinusoidal(sample_points, 1, 10) + make_sinusoidal(sample_points, 1, 15)
plt.title('fs = 20', fontsize = 15)
plt.stem(sample_points, y)
plt.show()

fs = 50
step = 1.0/fs
sample_points = np.arange(0, 2.0, step)
y = make_sinusoidal(sample_points, 1, 5) + make_sinusoidal(sample_points, 1, 10) + make_sinusoidal(sample_points, 1, 15)
plt.title('fs = 50', fontsize = 15)
plt.stem(sample_points, y)
plt.show()
```
