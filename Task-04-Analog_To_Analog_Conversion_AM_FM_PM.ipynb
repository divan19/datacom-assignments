```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
'''
t     = Time domain points for the signal
x_t   = Signal 
'''

def plot_signal(t, x_t):
  plt.figure(figsize=(18, 6))
  plt.plot(t, x_t)
  plt.show()
```

## **Amplitude Modulation(AM)**

**Carrier Signal**
$$c(t)=A_c cos(2 \pi f_c t)$$

**Message Signal**
$$m(t)=A_m cos(2 \pi f_m t)$$

**Amplitude modulated signal**
$$product = A_c * (1 + \mu * cos(2 \pi f_m t) * cos(2 \pi f_c t)$$


```python
'''
A_c  - Amplitude of carrier signal
f_c - Frequency of carrier signal
c_t - Carrier signal

A_m  - Amplitude of message signal
f_m - Frequency of message signal
m_t - Message signal
mu - modulation index
'''

A_c = 5
f_c = 4
A_m = 8
f_m = 6
mu = 3

t = np.linspace(0, 1, 1000)
c_t = A_c*np.cos(2*np.pi*f_c*t)
m_t = A_m*np.cos(2*np.pi*f_m*t)
product = A_c*(1+mu*np.cos(2*np.pi*f_m*t))*np.cos(2*np.pi*f_c*t)

# message signal
plot_signal(t, m_t)

# carrier signal
plot_signal(t, c_t)

# amplitude modulated signal
plot_signal(t, product)
```
